### MOVIE ANALYSIS

#### Overview
This analysis is for a film making company that wants to emerge best at the box office. This analysis will help this company to know what features of the TMDB database to put in consideration in order to be more profitable.This analysis will focus on Genres,Release_Date,Revenue,Budget,Production_Companies,Runtime columns.


#### Business Problem

- A film making company has not been performing well  at the box office. The company needs guidance on how to improve on their performance in order to emerge best at the box office.
- Indentify factors of success of the industry


#### Data Understanding

For this analysis I will be using TMDB_7000_Top_Rated_Movies and TMDB_10000_Popular_Movies datasets from Kaggle. The two datasets have a total of 44 columns.This is the link to Kaggle [link here](https://www.kaggle.com/datasets/ritayandhara/tmdb-dataset).

```python
# Import necessary libraries
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt

%matplotlib inline


import seaborn as sns
import matplotlib.ticker as ticker
import plotly.express as px
plt.style.use('fivethirtyeight')
```

### Cleaning the datasets

Removing all null values in the tmdb_df 

```python
# Getting the sum of null values
missing_values_count=tmdb_df.isna().sum()
missing_values_count
```

```python
# how many total missing values do we have?
total_cells = np.product(tmdb_df.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

*I prefer dropping the rows with null values because only 1.437% of my dataset is null values*

```python
# Dropping all the null valuesin the tmdb_df dataframe
tmdb_df.dropna(inplace=True)
```

```python
# Checking for duplicates in the tmdb_df dataframe
tmdb_df.duplicated().sum()
# tmdb_df has 7 duplicates i have to drop the duplicates before i start data analysis
# Dropping the duplicates
tmdb_df.drop_duplicates(inplace=True)
```

```python
# Finding the sum of null values in the tmdb_df1 dataframe
missing_values_count=tmdb_df1.isna().sum()
```

```python
# how many total missing values do we have?
total_cells = np.product(tmdb_df1.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100

```

```python
# the percentage of null value in the tmdb_df1 dataframe is 0.8903 which is barely a quarter of the dataframe we have lost a little bit of our dataset

# Drop all the null value in the tmdb_df1 dataframe
tmdb_df1.dropna(inplace=True)
```

```python
# Let me find out what is the number of duplicates in the tmdb_df1 dataframe
tmdb_df1.duplicated().sum()
# the tmdb_df1 has no duplicates
```

- Parsing the Release_Date into datetime format

```python
tmdb_df['Release_Date']=pd.to_datetime(tmdb_df['Release_Date'])
tmdb_df1['Release_Date']=pd.to_datetime(tmdb_df1['Release_Date'])
```

- Renaming columns so that all the column names will have uniformity

```python
# Renaming the country or origin column and the rating average column
tmdb_df= tmdb_df.rename(columns = {'Country_of_Origin': 'Country_Of_Origin','Rating_average':'Rating_Average'})
tmdb_df1= tmdb_df1.rename(columns = {'Country_of_Origin': 'Country_Of_Origin','Rating_average':'Rating_Average'})
```

- Merging two dataframes and checking for any duplicates

```python
# Merging the two datsets
df=pd.concat([tmdb_df,tmdb_df1])
# Checking for any duplicates in the new dataframe
df.duplicated().sum()
# The new dataframe has 5309 duplicates so i will drop all the duplicates
df.drop_duplicates(inplace=True)
# All the duplicates in df has been dropped
```

#### Data Analysis

- Exploring a new feature profit 

```python
# Exploring a new feature profit
df['Profit']=df['Revenue']-df['Budget']
```

```python
# Creating two more columns based on Release_Date
df['Release_Year'] = pd.to_datetime(df['Release_Date']).dt.year
df['Release_Month'] = pd.to_datetime(df['Release_Date']).dt.month
```

6.**Original Language**

```python
#  A piechart to show the distribution of original language

#Get value counts of original languages
original_language_counts = df['Original_Language'].value_counts()

# Get the top 5 languages
top_languages = original_language_counts[:5]

# Calculate the count for the "Others" category
others_count = original_language_counts[5:].sum()

# Create a new series with the top 5 languages and "Others"
languages_data = pd.concat([top_languages, pd.Series(others_count, index=['Others'])])

# Plotting 'original_language'
plt.figure(figsize=(12, 12))
languages_data.plot(kind='pie', autopct='%1.1f%%')
plt.title('Number of Movies by Original Language')
plt.ylabel('')

# Add labels to the pie chart
plt.legend(labels=languages_data.index)
plt.show()
```
